```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
```

```python
# conexión con la base de datos de mongo
client = MongoClient("mongodb://localhost:27017/",username='admin',password='123')
db = client.companies 
```

```python
# Query para pedir todo

query = {"$and":[ 
                {"founded_year": {"$gte": 2009}} ,    
                 {"offices.latitude":{"$ne": "null"}} ,
                 {"offices.longitud":{"$ne": "null"}} ,
                 {"total_money_raised": {"$regex" : "M" }} ,
    {"$or": [{"category_code": "design" } ,{"category_code": "web" } ,{"category_code": "software" } ,
    {"category_code": "games_video" } , {"category_code": "mobile" } , {"category_code": "enterprise" } , 
    {"category_code": "analytics" } ,{'category_code': "search"},{'category_code': "network_hosting"} ,
    {"category_code": "photo_video" } , ]} ,
                ]
        } 

objec = {"_id":0, "name":1, "total_money_raised":1, "founded_year":1, "offices.city":1,
         "offices.latitude":1,"offices.longitude":1}


companies = db.companies.find(query,objec)
```

```python
data_companies = pd.DataFrame(companies)
```

```python
def geopoint(offices):
    
    for i in offices:
        return {"type": "Point", "coordinates": [i['longitude'],i['latitude']]}

        #print(i['latitude'])
    #return offices
    #longitude = data_companies.offices[0][0]['longitude']
    #return latitude
```

```python
data_companies.offices[0][0]['latitude']
```

```python
data_companies['columnaNueva'] = data_companies['offices'].apply(geopoint)
data_companies.head()
```

```python

def geocity(offices):
    
    for i in offices:
        return i['city']
```

```python
data_companies['City'] = data_companies['offices'].apply(geocity)
data_companies.head()
```

```python
def geoclongitude(offices):
    
    for i in offices:
        return i['latitude']
```

```python
data_companies['latitude'] = data_companies['offices'].apply(geoclongitude)
data_companies
```

```python
def geoclongitude(offices):
    
    for i in offices:
        return i['longitude']
```

```python
data_companies['longitude'] = data_companies['offices'].apply(geoclongitude)
data_companies.head()
```
