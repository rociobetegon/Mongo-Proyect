```python
def starbucks(dictrandom):
    names=[]
    lats=[]
    long=[]
    for place in dictrandom["results"]:
        names.append(place["name"])
        lats.append(place["geometry"]["location"]["lat"])
        long.append(place["geometry"]["location"]["lng"])
    dictstarbucks = {"name":names, "latitude":lats, "longitude":long}
    return pd.DataFrame(dictstarbucks)
datastarbucks = starbucks(data)
datastarbucks.head()
```

```python
datastarbucks["City"] = "New York"
datastarbucks.head()
```

```python
def createGeoJson(office):
    return {
        "type":"Point",
        "coordinates":[office["longitude"],office["latitude"]]
    }
datastarbucks['position'] = datastarbucks.apply(createGeoJson, axis=1)

```
