```python
 # conexión con la base de datos de mongo
client = MongoClient("mongodb://localhost:27017/",username='admin',password='123')
db = client.companies 
```

```python
companiesdata = db.mongofinal.find()
```

```python
comp = pd.DataFrame(companiesdata)
comp
```

```python
def getCompaniesNear(latitude,longitude, max_meters=2000):
    return list(db.mongofinal.find({
        "geopoint": {
            "$near": {
               "$geometry": {
                  "type": "Point" ,
                  "coordinates": [ longitude , latitude ]
               },
               "$maxDistance": max_meters
             }
       }
    }))


fifth_avenue = [40.7744089,-73.9678246]
df_filtered = pd.DataFrame(getCompaniesNear(*fifth_avenue,6000))
df_filtered.count()
```

```python
result = df_filtered.drop(columns=["_id"])
result
```

```python
result.to_json('./mongoresult.json',orient="records")
```
